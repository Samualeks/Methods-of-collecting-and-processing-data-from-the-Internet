```python
from pymongo import MongoClient
```

```python
clothes = [ {'name': 'hat', 'price': 5},
    {'name': 'pant', 'price': 10},
    {'name': 'dress', 'price': 12},
    {'name': 'shirt', 'price': 7},
    {'name': 'tie', 'price': 3},
    {'name': 'jacket', 'price': 9} ]
```

```python
client = MongoClient('mongodb://localhost:27017/')
```

```python
print(client.list_database_names()) 
```

```python
db = client.testdb
```

```python
db.clothes.insert_many(clothes)
```

```python
print(db.list_collection_names())
```
